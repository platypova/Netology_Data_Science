```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import adjusted_rand_score, silhouette_score
import numpy as np
import matplotlib.pyplot as plt
```

### Загрузим данные

```python
df = pd.read_excel('geo.xlsx')
```

```python
df.head()
```

```python
X = df[['x','y']].values
```

### Проведем кластеризацию с помощью алгоритма K-means

```python
# Выберем количество кластеров = 10
clf_kmeans = KMeans(n_clusters=10)
clusters_kmeans = clf_kmeans.fit_predict(X, )
```

### Визуализируем кластеры ( районы голосований)

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1],c = clusters_kmeans)
plt.xlabel('Визуализация райнов голосования с помощью кластеризации координат голосований')
```

### Исследуем скопления голосов за и против отдельно

Голоса ЗА

```python
df_for = df[df['comment_class'] == 1] 
X_for = df_for[['x','y']].values
```

```python
# Выберем количество кластеров = 10
clf_kmeans = KMeans(n_clusters=10)
clusters_kmeans = clf_kmeans.fit_predict(X_for, )
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X_for[:,0], X_for[:,1],c = clusters_kmeans)
plt.xlabel('Визуализация райнов голосования с помощью кластеризации координат голосований')
```

Голоса ПРОТИВ

```python
df_against = df[df['comment_class'] == -1] 
X_ag = df_against[['x','y']].values
```

```python
# Выберем количество кластеров = 10
clf_kmeans = KMeans(n_clusters=10)
clusters_kmeans = clf_kmeans.fit_predict(X_ag, )
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X_ag[:,0], X_ag[:,1],c = clusters_kmeans)
plt.xlabel('Визуализация райнов голосования с помощью кластеризации координат голосований')
```

# Подберем оптимальное количество кластеров:

```python
k_inertia = []
ks = range(1,15)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
```

```python
diff = np.diff(k_inertia)
plt.plot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
```

```python
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```

Несмотря на то, что по этим вычислениям (взятым из лекционных материалов) мы получаем, что оптимальное число кластеров = 2, нам не подходит этот вариант. Возможно, с какой-то точки зрения это был бы подходящий вариант, но мы точно знаем, что районов Москвы больше, чем 2, поэтому нужно использовать другое число кластеров. Предложенное число 10 кажется подходящим для этой задачи.
